```python
import sys
sys.path.append("../src")
import os
import numpy as np
import os
from basic_run import lock_sindy_params, data_and_run
from training import load_model, TrainModel
import tensorflow as tf
import pickle
from lorenz import Lorenz, params2data
from sindy_utils import sindy_simulate, sindy_library_names
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from analyze import get_names
from tqdm import tqdm
import seaborn as sns

os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
```

```python
def get_checkpoint_names(filename):
    cpath = os.getcwd()+'/data/'+filename+'/checkpoints/'
    all_files = os.listdir(cpath)
    all_files = set([n.split('.')[0] for n in all_files])
    all_files.remove('checkpoint')
    all_files = list(all_files)
    all_files.sort()
    print('number of checkpoints = ', len(all_files))
    return all_files

def params2dict(params):
    params2 = {key: val[0] for key, val in params.to_dict().items()}
    list_to_int = ['input_dim', 'latent_dim', 'poly_order', 'n_ics', 'include_sine', 'exact_features']
    params2['lorenz_coefficients'] = list(params2['lorenz_coefficients'])
    for key in list_to_int:
        params2[key] = int(params2[key])
    
    return params2
```

```python
cases = ['data_size', 'lock_sindy', 'sindy_pert'] #['with_thresh', 'architecture_poly']
path = 'data/'


# file = 'results_202105030932_data_size_xz_x0_losses'
name_list = get_names(cases, path)[18:]
print(name_list)
```

```python
plt.rcParams["animation.html"] = "jshtml"
%matplotlib inline
```

```python
import pdb

for i in tqdm(range(len(name_list))):
    skip = False
    filename = name_list[i]
    print(filename)
    cpath = os.getcwd()+'/data/'+filename+'/checkpoints/'
    try:
        cp_files = get_checkpoint_names(filename)
    except:
        continue
        print('there arent any checkpoints')
    
    model, params, results = load_model(filename, os.getcwd()+'/data/')
    p = params2dict(params)
    p['tend'] = 30
    data = params2data(p)
    trainer = TrainModel(data, p)

    poly_order = p['poly_order']
    include_sine = p['include_sine']
    exact_features = p['exact_features']

    coef_names = sindy_library_names(p['latent_dim'], p['poly_order'], include_sine=False, exact_features=p['exact_features'])
    coef_names_full = sindy_library_names(p['latent_dim'], p['poly_order'], include_sine=False, exact_features=False)

    data.training_split(0.1)

    train_data, test_data = data.get_train_test()
    train_time, test_time = data.get_times()

    z0 = np.array([1/40, 1/40, 1/40])
    lorenz_sim = sindy_simulate(z0, test_time, data.sindy_coefficients, poly_order, include_sine)

    zl_list = []
    zs_list = []
    coef_list = []

    for i in range(len(cp_files)):
        name = cp_files[i]
        try:
            model.load_weights(cpath+name+'.ckpt').expect_partial()
        except:
            skip = True
            print('cannot load weights')
            print(name)
            break
        z_latent = model.encoder(test_data[0]).numpy()
        z_sim = sindy_simulate(z0, test_time, model.sindy.coefficients_mask* model.sindy.coefficients, 
                                   poly_order, include_sine, exact_features=exact_features)

        zl_list.append(z_latent)
        zs_list.append(z_sim)
        coef_list.append(model.sindy.coefficients.numpy())


    if skip:
        continue
    fig = plt.figure(figsize=(10, 10))
    ax3 = fig.add_subplot(111, projection='3d')
    ax3.plot(zl_list[0][:, 0], zl_list[0][:, 1], zl_list[0][:, 2], color = 'k', linewidth=1)
    ax3.legend(['latent z', 'simulated sindy', 'true z'])
    
    
    def animate(i):
        ax3.clear()
        ax3.plot(zl_list[i][:, 0], zl_list[i][:, 1], zl_list[i][:, 2], color = 'k', linewidth=1, alpha=0.5)
        ax3.plot(zs_list[i][:, 0], zs_list[i][:, 1], zs_list[i][:, 2], color = 'r', linewidth=0.5, alpha=0.5)
        ax3.plot(lorenz_sim[:, 0], lorenz_sim[:, 1], lorenz_sim[:, 2], color = 'g', linewidth=0.5, alpha=0.8)
        ax3.legend(['latent z', 'simulated sindy', 'true z'])
        ax3.set_xlabel('x')
        ax3.set_ylabel('y')
        ax3.set_zlabel('z')

        return [ax3]

    # run the animation
    animation = FuncAnimation(fig, animate, frames=len(zl_list), interval=50)
    # HTML(animation.to_html5_video())
    animation.save('./figures/'+ filename +'_animation_'+ '.mp4')

    titles = ['$\dot x$', '$\dot y$', '$\dot z$']
    true_coefs = p['actual_coefficients']
    coef_array = np.array(coef_list)
    nc = coef_array.shape[0]
    threshold = 1e-2
    cmap = sns.color_palette()

    fig = plt.figure(figsize=(20, 5))
    ax = []
    for i in range(3):
        a = fig.add_subplot(1, 3, i+1)
        leg = []
        for cidx, coef in enumerate(coef_names):
            tc = true_coefs[cidx, i]
            if abs(coef_array[-1, cidx, i]) > threshold:
                a.plot(np.arange(nc), coef_array[:, cidx, i], '.-', color=cmap[cidx])
                leg.append(coef_names[cidx])
            if abs(tc) > threshold:
                a.plot([0, nc], [tc, tc], '--', color=cmap[cidx])
                leg.append(coef_names[cidx])

        a.legend(leg)
        a.set_title(titles[i])
        a.set_xlabel('epoch')
        a.set_ylabel('coef value')
        ax.append(a)
    plt.savefig('./figures/'+ name +'_coefficients_'+ '.pdf')
    plt.show()
    # animation.save('example.mp4')
```
